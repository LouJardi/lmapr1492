```python

import pymatgen
import numpy as np
from pymatgen.io.cif import CifWriter
from pymatgen.ext.matproj import MPRester
from IPython.display import FileLink
from pymatgen.analysis.diffraction.xrd import XRDCalculator

```

```python
with MPRester("aeo5GUJ5RRpikp1p") as m:
    structure = m.get_structure_by_material_id("mp-961706")
    structure.to(filename = "TiSiPt.cif")
    writer = CifWriter(structure)
    writer.write_file("mp-961706.cif")
    
    print("Download .CIF file")
    display(FileLink("mp-961706.cif"))
    
xrd = XRDCalculator('CuKa')
```

```python
# Diffractogramme du TiSiPt et représentation des 3 premiers pics
pattern = xrd.get_pattern(structure)
xrd.show_plot(structure, two_theta_range=(1,50), annotate_peaks=True, ax=None, with_labels=True)
```

```python
# 1er pic. 
Angle = pattern.x[0]
Intens = pattern.y[0]
PdD = pattern.hkls[0]
DiP = pattern.d_hkls[0]


print("========================================================\n")
print("PREMIER PIC \n")
print("Angle 2𝜃 :                      ", Angle, " [°]")
print("Intensité du premier pic :      ", Intens, " [%] ")
print("Indices hkl des plans associés :", PdD[0], PdD[1])
print("Distance entre ces plans :      ", DiP, " [Å]\n")
print("========================================================")
```

```python
# 2ème pic.
Angle = pattern.x[1]
Intens = pattern.y[1]
PdD = pattern.hkls[1]
DiP = pattern.d_hkls[1]


print("========================================================\n")
print("SECOND PIC \n")
print("Angle 2𝜃 :                      ", Angle, " [°]")
print("Intensité du second pic :      ", Intens, " [%] ")
print("Indices hkl des plans associés :", PdD[0])
print("Distance entre ces plans :      ", DiP, "  [Å]\n")
print("========================================================")
```

```python


# 3ème Pic 
Angle = pattern.x[2]
Intens = pattern.y[2]
PdD = pattern.hkls[2]
DiP = pattern.d_hkls[2]

print("========================================================\n")
print("TROISIEME PIC \n")
print("Angle 2𝜃 :                      ", Angle, " [°]")
print("Intensité du 3ème pic :         ", Intens, " [%] ")
print("Indices hkl des plans associés :", PdD[0], PdD[1])
print("Distance entre ces plans :      ", DiP, " [Å]\n")
print("========================================================")

```
